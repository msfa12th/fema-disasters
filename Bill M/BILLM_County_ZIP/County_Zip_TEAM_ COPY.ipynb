```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Location/zip_by_county.csv"
county_zip_df = pd.read_csv(csv_file)
county_zip_df.head()
```

```python
new_countzip_df = county_zip_df[["ID","ZIP", "STCOUNTYFP", "CITY", "STATE", "COUNTYNAME", "CLASSFP"]].copy()
new_countzip_df.head()
```

```python
rds_connection_string = "<DB USER>:<ENTER DB PASSWORD>@localhost:5432/<ENTER DATABASE NAME>"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_countzip_df = new_countzip_df.rename(columns = {"ID": "id",
                                    "ZIP": "zip", 
                                  "STCOUNTYFP":"st_county_fp", 
                                  "CITY": "city",
                                      "STATE": "state",
                                      "COUNTYNAME": "county_name",
                                      "CLASSFP": "class_fp"})

new_countzip_df.head()
```

```python
new_countzip_df.to_sql(name='county_zip', con=engine, if_exists='append', index=False)
```
