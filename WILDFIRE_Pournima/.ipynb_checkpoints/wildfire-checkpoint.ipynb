This application reads data from multiple files 
(original large file split into smaller files for github storage)
Then extracts necessary columns, transforms values as needed and loads into wildfire table for posterity
Finally (via SQL script) it loads common fields into NATURAL_DISASTER table for easy comparison with other source data.
Original source: https://catalog.data.gov/dataset/fire-program-analysis-fire-occurrence-database-feature-layer

```python
import pandas as pd
from sqlalchemy import create_engine

import glob
import os
```

```python
#get list of files
filenames = []
path =r'wildfiredividedrawdata/'
with os.scandir(path) as entries:
    for entry in entries:
        filenames.append(entry.name)
```

```python
filenames.remove('.DS_Store')
```

```python
filenames.sort()
```

```python
#concatenate all files into single big file
with open('wildfireOutput.csv', 'w') as outfile:
    for fname in filenames:
        fpName = path + fname
        with open(fpName) as infile:
            for line in infile:
                outfile.write(line)
```

```python
csv_file = "wildfireOutput.csv"
wildfire_data_df = pd.read_csv(csv_file,dtype='unicode')
wildfire_data_df.columns
```

```python
cols = []
for col in wildfire_data_df.columns:
    cols.append(col.lower())
wildfire_data_df.columns = cols
cols
```

```python
#selector = wildfire_data_df['stat_cause_descr'] == 'Lightning'
#lightning_data_df = wildfire_data_df.loc[ selector ].copy()
wildfire_data_df.drop(['x','y'],axis = 1)
wildfire_data_df['data_source'] = 'WILDFIRE'
#wildfire_data_df['data_source'] = 1
wildfire_data_df['id'] = wildfire_data_df.index
wildfire_data_df.head().T
```

```python
wildfire_drop_data_df= wildfire_data_df.drop(["x","y"], axis = 1)
```
