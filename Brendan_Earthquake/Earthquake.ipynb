```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "brandensfile.csv"
eq_df = pd.read_csv(csv_file)
eq_df.head()
```

```python
new_eq_df = eq_df[["EarthQ_ID", "ID", "Date", "Time", "Latitude", "Longitude", "Type", "Depth", "Magnitude", "Magnitude Type", "Source", "Status"]].copy()
new_eq_df.head()
```

```python
rds_connection_string = "<DB USER>:<ENTER DB PASSWORD>@localhost:5432/<ENTER DATABASE NAME>"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_eq_df = new_eq_df.rename(columns = {
                                  "EarthQ_ID": "earthq_id",
                                   "ID": "id",
                                  "Date": "date", 
                                  "Time":"time", 
                                  "Latitude": "latitude",
                                   "Longitude": "longitude",
                                   "Type":"disaster_type",
                                   "Depth":"depth",
                                   "Magnitude":"magnitude",
                                   "Magnitude Type":"magnitude_type",
                                   "Source":"data_source",
                                   "Status":"status"
                                         })
                                      

new_eq_df.head()

new_eq_df.to_sql(name='earthquake', con=engine, if_exists='append', index=False)
```
