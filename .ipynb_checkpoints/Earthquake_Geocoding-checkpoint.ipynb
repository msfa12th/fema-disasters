```python
# Import the necessary libraries
# Original g_key no longer active
import pandas as pd
import requests
import json
from api_keys import g_key
```

```python
# Import the Earthquake CSV
EQ_df = pd.read_csv("Resources/Earthquake_v1.csv")
EQ_df = EQ_df.drop(['Depth Error', 'Depth Seismic Stations', 'Magnitude Error', 'Magnitude Seismic Stations', 'Azimuthal Gap', 'Horizontal Error', 'Horizontal Distance', 'Root Mean Square', 'Location Source', 'Magnitude Source',], axis=1)
EQ_df['ZipCode'] = ""
EQ_df['City'] = ""
EQ_df['State'] = ""

EQ_df.head()
```

### REMEMBER TO REIMPORT FROM CSV IF BELOW DROPPING FUNCTION NEEDS TO BE REDONE

```python
count = 0

# Reduce the Indexes to just those concerning the USA (rough box around) via Latitude and Longitude checks

for index, row in EQ_df.iterrows():
    
    lat = row[2]
    lng = row[3]
    if (20 < lat <= 50 and -130 <= lng <= -60) \
    or (50 < lat <= 75 and -180 <= lng <= -130):
        count += 1
    else:
        EQ_df = EQ_df.drop(index)
print(count, len(EQ_df))
```

```python
# For Loop to append location data to EQ_df for values in USA Federal States
EQ_df = EQv2_df
params = {'key': g_key}

base_url = "https://maps.googleapis.com/maps/api/geocode/json"

for index, row in EQ_df.iterrows():
    lat = row['Latitude']
    lng = row['Longitude']
    
    params['latlng'] = f'{lat},{lng}'
    
    result = requests.get(base_url, params=params).json()
    
    try:
        address = result['results'][0]['formatted_address'].split(',')
        country = address[3].strip(' ')
    
        if country == ' USA':
            EQ_df.loc[index, 'ZipCode'] = address[2].split(' ')[-1]
            EQ_df.loc[index, 'City'] = address[1].lstrip(' ')
            EQ_df.loc[index, 'State'] = address[2].split(' ')[1]
        else:
            pass
            
    except:
         pass

print(len(EQ_df))
```
