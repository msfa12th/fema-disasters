```python
import pandas as pd
from sqlalchemy import create_engine
import os
```

```python
csv_file = os.path.join("tornadorawdata.csv")
tornadoraw_data_df = pd.read_csv(csv_file)
tornadoraw_data_df
```

```python
new_tornadoraw_data_df = tornadoraw_data_df[["id", "om", "data_source","yr", "mo", "dy", "date", "tz", "st", "stf", "stn",
         "mag", "inj", "fat", "loss","closs", "slat", "slon", "elat", "elon", "len", "wid","ns", "sn", "sg", "f1", "f2",
        "f3", "f4", "fc"]].copy()


new_tornadoraw_data_df
```

```python
rds_connection_string = "postgres:Dbadmin@localhost:5432/NaturalDisaster_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```
